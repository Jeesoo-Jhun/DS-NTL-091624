### 3b: What are some of the things we could do to deal with null values?
 

```python
# call_on_students(1)

# We could:
# 1. Drop rows with any null values using df.dropna().
# 2. Fill null values with a specific value using df.fillna(value).
# 3. Fill null values with the mean or median of the column using df.fillna(df['column'].mean()) or df.fillna(df['column'].median()).
# 4. Use forward fill or backward fill methods using df.fillna(method='ffill') or df.fillna(method='bfill').

#in this case, how would you handle the missing values?

# In this case, I would handle the missing values as follows:
# 1. For the 'Paid' column, I would fill the missing value with 0, assuming that if it's not specified, the post was not paid.
df['Paid'].fillna(0, inplace=True)

# 2. For the 'like' column, I would fill the missing value with the median of the column, as likes can vary widely and the median is a good central tendency measure.
df['like'].fillna(df['like'].median(), inplace=True)

# 3. For the 'share' column, I would drop the rows with missing values, as shares are crucial for engagement metrics and we have a few missing values.
df.dropna(subset=['share'], inplace=True)
```

    - replace 0 or mean

```python
# Your code here

def find_song_by_artist(artist_name, song_dict):
    
    '''
    Parameters:
    artist_name: a string of an artist's name to be used to search the dictionary
    song_dict:  a dictionary of top_twenty songs with song name as keys and a list of 
    artist names as values
    
    Returns:
    A list of songs which the given artist appeared on
    '''
    song_list = []
    for song, artist_tuple in song_dict.items():
        if artist_name in artist_tuple:
            song_list.append(song)
    return [song for song, artists in song_dict.items() if artist_name in artists]
```
